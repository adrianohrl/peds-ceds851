```python
def count_words(x):
    return 0 if x is np.nan else len(x.split())

def get_numbers(t):
    num_dic = {'sete': 7, 
               'dez': 10, 
               'quinze': 15, 
               'cinco': 5, 
               'trinta': 30, 
               'vinte': 20, 
               'três': 3, 
               'sessenta': 60, 
               'noventa': 90}
    num = np.nan
    if re.search(r'o prazo de (.*?) dias', t) is not None:
        r = re.search(r'o prazo de (.*?) dias', t).group(1).lower()
        if len(r) < 25:
            lst = re.findall('\d+', r)
            num = num_dic.get(r) if len(lst) == 0 else max(lst)
    if str(num)[:1] == 0:
        num = int(str(num[1:]))
    return num
```

```python
def preprocessor(txt, stop_words = nltk.corpus.stopwords.words('portuguese'), stemmer = nltk.stem.RSLPStemmer(), stem = True):
    if type(txt) != str:
        return txt
    txt = txt.lower()
    
    txt = ' '.join([word for word in txt.split() if word not in stop_words])
    if stem:
        txt = ' '.join([stemmer.stem(word) for word in txt.split()])
#     txt = re.sub(r'[ãáà]', 'a', txt)
#     txt = re.sub(r'[éê]', 'e', txt)
#     txt = re.sub(r'[í]', 'i', txt)
#     txt = re.sub(r'[õó]', 'o', txt)
#     txt = re.sub(r'[ú]', 'u', txt)
#     txt = re.sub(r'[ç]', 'c', txt)
#     #txt = re.sub(r'[\W\d]',' ', txt)
    txt = re.sub(r'\s+',' ', txt)
    return txt  
```

```python
def _split(X, y, train_indexes, test_indexes, column = 'process'):
    train_indexes = X[column].isin(train_indexes)
    test_indexes = X[column].isin(test_indexes)
    return X.loc[train_indexes], X.loc[test_indexes], y.loc[train_indexes], y.loc[test_indexes]

def split(X, x, split_column, index_column = 'process', target_column = 'delta', test_size = 0.2, random_state = 44):
    '''
    Input Parameters:
        X: original dataframe;
        x: simplified dataframe (grouped by index_column);
        index_column: column in which the simplified dataframe were grouped by; 
        target_column: the target column in the original dataframe;
        test_size: the split proportion (see description in the train_test_split function);
        random_state: (see description in the train_test_split function).
    '''
    y = x[split_column]
    if y.dtype in [pd.np.dtype('float64'), pd.np.dtype('float32')] or np.min(np.bincount(y)) < 2:
        y = np.digitize(y, range(10))    
    X_train, X_test = train_test_split(x.loc[:, x.columns != split_column], stratify = y, test_size = test_size, random_state = random_state)   
    return _split(X.loc[:, X.columns != target_column], X[target_column], X_train[index_column], X_test[index_column], index_column)
```

### Model

```python
LSTM_SIZE = 20
DENSE1_SIZE = 256
DENSE2_SIZE = 128
LEARNING_RATE = 0.0001
RHO = 0.9
EPSILON = None
DECAY = 0.0

main_input = Input(shape = (X_train_main.shape[1], ), dtype = 'int32', name = 'main_input')

x = Embedding(MAX_WORDS, EMBEDDING_DIM, 
              input_length = MAX_LEN, 
              weights = [embedding_weights], 
              name = 'embedding',
              trainable = False)(main_input)

lstm_out = LSTM(LSTM_SIZE, name = 'lstm')(x)

aux_input = Input(shape = (X_train_aux.shape[1], ), name = 'aux_input')

x = concatenate([lstm_out, aux_input])
x = Dense(DENSE1_SIZE, kernel_initializer = 'uniform', activation = 'relu', name = 'dense1')(x)
x = Dense(DENSE2_SIZE, kernel_initializer = 'uniform', activation='relu', name = 'dense2')(x)

main_output = Dense(1, activation = 'linear', kernel_initializer = 'normal', name = 'main_output')(x)

model = Model(inputs = [main_input, aux_input], outputs = [main_output])
optimizer = keras.optimizers.RMSprop(lr = LEARNING_RATE, rho = RHO, epsilon = EPSILON, decay = DECAY)
model.compile(optimizer = optimizer, loss = 'mse', metrics = ['mse'])
model.summary()
```
